```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取 CSV 文件（请根据实际路径调整）
csv_path = "./lora_sdxl_inpainting3/training_loss.csv"
df = pd.read_csv(csv_path)

# 按 global_step 分组，每200步为一组（例如0-99, 100-199, ...）
df['step_group'] = (df['global_step'] // 300) * 300

# 计算每组的平均损失
grouped_loss = df.groupby('step_group')['loss'].mean().reset_index()

# 绘制折线图
plt.figure(figsize=(8,6))
plt.plot(grouped_loss['step_group'], grouped_loss['loss'], marker='o')
plt.xlabel('Global Step (per300steps)')
plt.ylabel('average loss')
plt.title('average losses/300steps')
plt.grid(True)
plt.show()
```
